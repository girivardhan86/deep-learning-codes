```
from sklearn.datasets import make_classification
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate linear data
X, y = make_classification(
    n_samples=200,     # total samples
    n_features=2,      # only 2 features (for easy plotting)
    n_redundant=0,     # no redundant features
    n_informative=2,   # both features are informative
    n_clusters_per_class=1,
    class_sep=2.0,     # higher = easier separation
    random_state=42
)

# Convert to DataFrame for visualization
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Class'] = y

# Plot
sns.scatterplot(x='Feature1', y='Feature2', hue='Class', data=df, palette='coolwarm', s=80)
plt.title("Linearly Separable 2-Class Data")
plt.show()
```

```
X=df.iloc[:,0:2]
y=df.iloc[:,2]
```

```
from sklearn.linear_model import Perceptron
# Assuming you already have X and y (2-class linear data)
clf = Perceptron(max_iter=1000, eta0=0.1, random_state=42)
clf.fit(X, y)
```

```
print(clf.coef_)
clf.intercept_
```

```
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# X and y are already NumPy arrays
plot_decision_regions(X, y, clf=clf, legend=2)  # clf is your trained Perceptron
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Perceptron Decision Regions")
plt.show()
```

```
# X and y are your training data
accuracy = clf.score(X, y)
print("Training accuracy:", accuracy)
```
