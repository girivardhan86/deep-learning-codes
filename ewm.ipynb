```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df=pd.read_csv('/content/DailyDelhiClimateTest.csv')
```

```python
df=df[['date','meantemp']]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['date'], df['meantemp'], color='red')
plt.xlabel("date")
plt.ylabel("meantemp")
plt.show()
```

```python
x1=df.iloc[:,1:2].ewm(alpha=0.9).mean()
```

```python
df['ewm']=x1
```

```python
df.head()
```

```python

plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'],x1,color='black',label='alpha=0.9')
plt.xlabel("date")
plt.ylabel("meantemp")
plt.legend()
plt.show()
```
